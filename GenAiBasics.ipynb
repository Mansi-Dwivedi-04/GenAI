### Using Transformer getting embeddings

```python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

# Input sentence
sentence = "you are mine"
tokenized_val = tokenizer(sentence,return_tensors='pt')

# This tokenized_val is vector representation of our sentence. 

#Get Embedding
with torch.no_grad():
    outputs = model(**tokenized_val)
```

```python
cls_embedding = outputs.last_hidden_state[:, 0, :]
cls_embedding
```

## Write a langchain chain to tell the capital of Hungary

```python
from langchain.chat_models import init_chat_model
from langchain_core.prompts import PromptTemplate
from langchain.chains import LLMChain

# Initialize LLM
LLM = init_chat_model("llama3-8b-8192",model_provider="groq")

# Create a prompt template
prompt = PromptTemplate(
    input_variable = ['question'],
    template= "\n\n{question}"
)

# Create a chain
chain = LLMChain(llm=LLM,prompt=prompt)

#Ask a ques
question = "What is the capital of Antartica?"
result = chain.run(question)

print(result)
```

### You are building a GenAI app that can answer questions about a company’s internal documents. Instead of fine-tuning a model, you decide to use LangChain with Retrieval-Augmented Generation (RAG).

**Your task:**
Build a mini LangChain RAG pipeline that does the following:

* Load a small text document.

* Index them using a vector store .

* Use ChatGroq to answer a user query using retrieved context.

* Return the generated answer using LangChain’s RetrievalQA.

```python
import bs4
from langchain import hub
from langchain.chat_models import init_chat_model
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langgraph.graph import START,END,StateGraph
from langchain_community.document_loaders import WebBaseLoader
from typing_extensions import List, TypedDict
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate

# Load Chat and Embedding Model 
llm = init_chat_model("llama3-8b-8192", model_provider="groq")
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
vector_store = InMemoryVectorStore(embeddings)

#1. Document
loader = WebBaseLoader(
    web_paths=("https://medium.com/@haroldfinch01/what-is-ansible-galaxy-d64ad8cf72fc",),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("ci bh gb gc gd ge")
        )
),
)

docs = loader.load()

#2. Splitting
text_splitters = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200)

all_splits = text_splitters.split_documents(docs)


#3. Indexing
_ = vector_store.add_documents(documents=all_splits)


# Method 1
prompt = hub.pull("rlm/rag-prompt",api_url="https://api.smith.langchain.com")

class State(TypedDict):
    question: str
    context: List[Document]
    answer: str

def retrieve(state: State):
    retrieved_docs = vector_store.similarity_search(state['question'])
    return {"context": retrieved_docs}

def generate(state: State):
    docs = "\n\n".join(doc.page_content for doc in state["context"])
    messages = prompt.invoke({"question":state["question"],
                             "context":docs})
    response = llm.invoke(messages)
    return {"answer":response}

graph_builder = StateGraph(State).add_sequence([retrieve,generate])
graph_builder.add_edge(START,"retrieve")

graph = graph_builder.compile()

response = graph.invoke({"question":"What is Ansible Galaxy?"})
response["answer"].content
```

```python
# Method 2

system_prompt = (
    "Use the given context to answer the question. "
    "If you don't know the answer, say you don't know. "
    "Use three sentence maximum and keep the answer concise. "
    "Context: {context}"
)
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}"),
    ]
)
retriever=vector_store.as_retriever()
question_answer_chain = create_stuff_documents_chain(llm, prompt)
chain = create_retrieval_chain(retriever, question_answer_chain)

response = chain.invoke({"input": "What is Ansible Galaxy"})
response['answer']
```
